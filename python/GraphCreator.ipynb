#GraphCreator

* Scrapping and loading the nodes (train stations)

```python
import gpxpy
import urllib2 
import networkx as nx
import unicodedata
import matplotlib.pyplot as plt
%matplotlib inline


data = urllib2.urlopen("http://datos.madrid.es/datosabiertos/BDC/POIS_TRANSPORTE/Metro/2015/Metro_2015_02.gpx")
gpx = gpxpy.parse(data)
#new Graph
Gr=nx.Graph()

for wpt in gpx.waypoints:
    name = wpt.name.split(" ")
    Gr.add_node(" ".join(name[1:]),line=name[:1],latitude=wpt.latitude, longitude= wpt.longitude)

#print the whole graph
nx.draw(Gr)
```

* Scrapping the edges and adding them to the graph

```python
from BeautifulSoup import BeautifulSoup
import urllib2
import unicodedata
from itertools import izip, islice


def createDoubleList(inputList):
    doubleList = []
    for current_item, next_item in izip(inputList, islice(inputList, 1, None)):
        doubleList.append([current_item, next_item])
    return doubleList


urlBase = 'https://www.metromadrid.es/es/viaja_en_metro/red_de_metro/lineas_y_horarios/linea'
for i in range(1,12):
    url=  urlBase + '%02d' % i + '.html'
    req = urllib2.Request(url,headers={'User-agent' : 'Mozilla/5.0'})
    con = urllib2.urlopen(req)
    soup = BeautifulSoup(con)
    tmpList= []
    for row in soup.html.body.findAll('tr'):
        data = row.findAll('a')
        for elem in data:
            if(elem.text!=""):
                tmpList.append(unicodedata.normalize('NFKD', elem.text).encode('ASCII', 'ignore').upper())
    tmpList = createDoubleList(tmpList)
    for first,second in tmpList:
        Gr.add_edge(first,second)
```

* Resulting graph

```python
#print the whole graph
nx.draw(Gr)
```

* Save graph to file

```python
nx.write_gpickle(Gr,"metro.gefx")
```
