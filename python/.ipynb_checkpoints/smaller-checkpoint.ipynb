```python
import networkx as nx
import matplotlib.pyplot as plt
import itertools
%matplotlib inline

Gr = nx.Graph()
Gr.add_node("A",mode="normal")
Gr.add_node("B",mode="normal")
Gr.add_node("C",mode="normal")
Gr.add_node("D",mode="normal")

Gr.add_edge("A","C",line="green")
Gr.add_edge("C","D",line="green")
Gr.add_edge("B","C",line="blue")
Gr.add_edge("C","D",line="red")

edges = Gr.edges()
colors = [Gr[u][v]["line"] for u,v in edges]
nx.draw(Gr,edges = edges, edge_color=colors,with_labels=True)
```

```python
for node in Gr.degree_iter():
    if(node[1]>2) & (Gr.node[node[0]]["mode"]!="purple"):
        dict = Gr[node[0]]
        newNodes = []
        for elem in dict.keys():
            nodeName = elem+node[0]
            print(nodeName)
            newNodes.append(nodeName)
            Gr.add_node(nodeName)
            Gr.add_edge(nodeName,elem,line=dict[elem]["line"])
        combinations = itertools.combinations(newNodes, 2)
        for elem in combinations:
            Gr.add_edge(elem[0],elem[1],line="purple")
            print("should be added "+"".join(elem))
        Gr.remove_node(node[0])
```

```python
edges = Gr.edges()
colors = [Gr[u][v]["line"] for u,v in edges]
nx.draw(Gr,edges = edges, edge_color=colors,with_labels=True)
```
